## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


````python
# 1. How many listings are there with a price less than 100? 

def get_listings_less_than_price(price):
    return listings_df['price'] < price
    
result = get_listings_less_than_price(100).sum()

print(f'Listings: {result}')
````

````python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

def get_listings_by_borough(borough):
    results = listings_df['neighbourhood_group'] == borough
    return listings_df[results]
    
    
df_bk = get_listings_by_borough('Brooklyn')
listings_count = len(df_bk)

print(f'Brooklyn listings: {listings_count}')
````

````python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

brooklyn_listings = get_listings_by_borough('Brooklyn')
filtered_listings = len(brooklyn_listings[get_listings_less_than_price(100)])

print(f'Listings: {filtered_listings}')
````

````python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

names = ['Michael', 'David', 'John', 'Daniel']
condition = listings_df['host_name'].isin(names)
result = len(listings_df[condition])

print(f'Total rows: {result}')
````

````python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

def adjust_price(listing):
    if listing['neighbourhood'] == 'Williamsburg':
        return listing['price'] + 100
    else:
        return listing['price']
    
    
listings_df['adjusted_price'] = listings_df.apply(adjust_price, axis=1)
listings_df.head(25)
````

````python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

def get_room_type_rows(room_type):
    room_condition = listings_df['room_type'] == room_type
    return len(listings_df[room_condition])
    

def calc_percentage(row):
    total_rows = len(listings_df)
    return (row / total_rows) * 100


private_room_rows = get_room_type_rows('Private room')
shared_room_rows = get_room_type_rows('Shared room')

percentage_private_rooms = calc_percentage(private_room_rows)
percentage_shared_rooms = calc_percentage(shared_room_rows)

print(f'Private Rooms: {percentage_private_rooms:.2f}%\nShared Rooms: {percentage_shared_rooms:.2f}%')
````

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

````python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

gb_neighborhood = listings_df.groupby('neighbourhood_group')
neighborhood_counts = gb_neighborhood.size()

print(f'{neighborhood_counts}')
````

````python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

gb_neighborhood['price'].mean()
````

````python
# 2.5. Using `groupby`, find the mean price for each room_type. 

gb_room_type = listings_df.groupby('room_type')
gb_room_type['price'].mean()
````

````python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

gb_neighborhood['price'].agg(['min', 'max'])
````

````python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.

mean_price = listings_df.groupby(['neighbourhood_group', 'room_type'])['price'].mean()
print(mean_price)
````

````python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

price_stats = listings_df.groupby(['neighbourhood_group', 'room_type']).agg(
    count=('price', 'count'),
    min=('price', 'min'),
    max=('price', 'max'),
    mean=('price', 'mean'),
    median=('price', 'median'),
    std=('price', 'std')
)

print(price_stats)
````

# Join and file saving.

````python
# 1. Load the `prices.csv` and the `n_listings.csv`

prices_df = pd.read_csv('data/prices.csv')
listings_df = pd.read_csv('data/n_listings.csv', sep=';')
````

````python
# 2. Do join that keeps all the records for each table.

joined_df = pd.merge(prices_df, listings_df, on='neighbourhood_group', how='outer')

valid_neighborhoods = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'LongIsland']
filtered_joined_df = joined_df[joined_df['neighbourhood_group'].isin(valid_neighborhoods)]
````

````python
# 3. Save your joined csv as 'joined.csv'.

filtered_joined_df.to_csv('data/joined.csv', index=False)
````

### 3. (Medium) Use the grammys.csv data for the next section of questions.

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?


3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

````python
# Setup dataframe

grammys_df = pd.read_csv('data/grammys.csv')
````

````python
# 1. Who has won Album of the Year in 2016?

def find_winner(category, year):
    winner = grammys_df[
        (grammys_df['year'] == year) & 
        (grammys_df['category'] == category) & 
        (grammys_df['winner'])
    ]
    
    if not winner.empty:
        workers = winner.iloc[0]['workers']
        artist = workers.split(',')[0].split('(')[0].strip()
        return artist
    else:
        return 'No matching record found'
    
    
print(f'Winning Artist: {find_winner("Album of the Year", 2016)}')
````

````python
# 2. Who won Best Rap Album in 2009?

print(f'Winning Artist: {find_winner("Best Rap Album", 2009)}')
````

````python
# 3. How many awards was Kendrick Lamar nominated for, and how many did he win?

def find_nominations(nominee):
    nominations = grammys_df[
        (grammys_df['nominee'] == nominee) |
        (grammys_df['workers'].str.contains(nominee))
    ]
    
    return nominations

def find_awards_won(nominations):
    awards_won = nominations[nominations['winner'] == True]
    return len(awards_won)
    

filtered_nominees = find_nominations('Kendrick Lamar')
nominated_count = len(filtered_nominees)

awards_count = find_awards_won(filtered_nominees)

print(f'Awards nominated for: {nominated_count}\nAwards won: {awards_count}')
````

### (Hard) Load the Game Logs for 2022 and add the column names using a dictionary.

````python
def create_dictionary():
    num_keys = 161
    blank_dict = {i: "" for i in range(1, num_keys + 1)}
    
    with open('data/game_logs/gamelog_columns.txt', 'r') as file:
        for i, line in enumerate(file, start=1):
            if i > num_keys:
                break
            blank_dict[i] = line.strip()
    return blank_dict


column_dict = create_dictionary()

game_log_2022_df = pd.read_csv('data/game_logs/gl2022.txt', header=None)
game_log_2022_df.columns = range(1, len(game_log_2022_df.columns) + 1)
game_log_2022_df.rename(columns=column_dict, inplace=True)

game_log_2022_df.head()
````

### (Extra Hard) Download the files for the past 5 years into a new folder and add them all into one data frame using pandas, then save that new file.
* Try to not hard code in the file names. We want to do this programmatically because what if we want to add new/more file names in the future.

````python
import os
import glob


data_dir = 'data/game_logs'
file_pattern = os.path.join(data_dir, 'gl*.txt')
file_paths = glob.glob(file_pattern)
file_paths.sort()

game_logs_df = []

for file_path in file_paths:
    temp_df = pd.read_csv(file_path, header=None)
    temp_df.columns = range(1, len(temp_df.columns) + 1)
    game_logs_df.append(temp_df)
    
combined_game_logs_df = pd.concat(game_logs_df, ignore_index=True)
combined_game_logs_df.rename(columns=column_dict, inplace=True)

combined_game_logs_df.head()
````
